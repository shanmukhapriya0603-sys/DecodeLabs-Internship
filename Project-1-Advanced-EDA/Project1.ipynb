```python
import pandas as pd

df = pd.read_excel("Dataset for Data Analytics (3).xlsx")

df.head()
```

```python
df.isnull().sum()
```

```python
df["CouponCode"] = df["CouponCode"].fillna("No Coupon")
```

```python
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df[(df["TotalPrice"] >= lower_bound) & (df["TotalPrice"] <= upper_bound)]

df_clean.shape
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df["TotalPrice"]))

df_z = df[z_scores < 3]
df_z.shape
```

```python
df["PricePerItem"] = df["TotalPrice"] / df["Quantity"]
```

```python
df["HighValueOrder"] = df["TotalPrice"].apply(lambda x: 1 if x > df["TotalPrice"].mean() else 0)
```

```python
def size_category(q):
    if q <= 2:
        return "Small"
    elif q <= 5:
        return "Medium"
    else:
        return "Large"

df["OrderSizeCategory"] = df["Quantity"].apply(size_category)
```
